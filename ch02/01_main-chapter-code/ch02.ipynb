## 2.3 토큰을 토큰 ID로 변환하기

- 그다음 텍스트 토큰을 나중에 임베딩 층에서 처리할 수 있는 토큰 ID로 변환합니다.

<img src="https://sebastianraschka.com/images/LLMs-from-scratch-images/ch02_compressed/06.webp" width="700px">

- 고유한 토큰으로 구성된 어휘사전을 만듭니다.

````python
all_words = sorted(set(preprocessed)) # preprocessed에서 고유한 단어를 추출하여 정렬합니다.
vocab_size = len(all_words)

print(vocab_size)
````

````python
vocab = {token:integer for integer,token in enumerate(all_words)}
````

- 어휘사전에 있는 처음 50개 항목을 확인합니다:

````python
for i, item in enumerate(vocab.items()):
    print(item)
    if i >= 50:
        break
````

- 아래 그림은 작은 어휘사전을 사용해 샘플 텍스트를 토큰화하는 과정을 보여줍니다:

<img src="https://sebastianraschka.com/images/LLMs-from-scratch-images/ch02_compressed/07.webp?123" width="700px">

- 이를 모두 합쳐서 토큰화 클래스를 만듭니다.

````python
class SimpleTokenizerV1:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {i:s for s,i in vocab.items()}

    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text) # 'hello,. world'

        preprocessed = [
            item.strip() for item in preprocessed if item.strip()
        ]
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids

    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        # 구둣점 문자 앞의 공백을 삭제합니다.
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
        return text
````

- `encode` 함수는 텍스트를 토큰 ID로 바꿉니다.
- `decode` 함수는 토큰 ID를 텍스트로 바꿉니다.

<img src="https://sebastianraschka.com/images/LLMs-from-scratch-images/ch02_compressed/08.webp?123" width="700px">

- 토크나이저를 사용해 텍스트를 정수로 인코딩(즉, 토큰화)합니다.
- 이 정수를 (나중에) 임베딩하여 LLM의 입력으로 사용할 수 있습니다.

````python
tokenizer = SimpleTokenizerV1(vocab)

text = """"It's the last he painted, you know,"
           Mrs. Gisburn said with pardonable pride."""
ids = tokenizer.encode(text)
print(ids)
````

- 정수를 텍스트로 다시 디코딩할 수 있습니다.

````python
tokenizer.decode(ids)
````

````python
tokenizer.decode(tokenizer.encode(text))
````

## 2.4 특수 문맥 토큰 추가하기

- 알지 못하는 단어와 텍스트의 끝을 알리는 "특수" 토큰을 추가하면 도움이 됩니다.

<img src="https://sebastianraschka.com/images/LLMs-from-scratch-images/ch02_compressed/09.webp?123" width="700px">

- 일부 토크나이저는 특수 토큰을 사용해 LLM이 추가적인 맥락을 이해하도록 돕습니다.
- 이런 특수 토큰에는 다음과 같은 것들이 있습니다.
  - `[BOS]` (beginning of sequence)는 텍스트의 시작을 표시합니다
  - `[EOS]` (end of sequence)는 텍스트의 끝을 표시합니다(서로 다른 두 개의 위키백과 문서나 책과 같이 관련이 없는 여러 텍스트를 연결할 때 사용합니다.)
  - `[PAD]` (padding) 하나 이상의 배치 크기로 LLM을 훈련할 때 배치 안에 길이가 다른 텍스트가 포함될 수 있습니다. 모든 텍스트의 길이를 동일하게 맞추기 위해 짧은 텍스트를 패딩 토큰을 사용해 배치에서 가장 긴 텍스트의 길이까지 확장합니다.
- `[UNK]`는 어휘 사전에 없는 단어를 나타냅니다.
- GPT-2는 위와 같은 토큰이 필요 없으며 복잡도를 줄이기 위해 `<|endoftext|>` 토큰만 사용합니다.
- `<|endoftext|>`는 위에서 언급한 `[EOS]` 토큰과 같습니다.
- GPT는 `<|endoftext|>`를 패딩에도 사용합니다(배치 입력에서 모델을 훈련할 때 마스크를 사용하기 때문에 패딩된 토큰에는 주의를 기울이지 않아 어떤 토큰을 사용하던지 관계가 없습니다).
- GPT-2는 어휘사전에 없는 토큰을 위해 `<UNK>` 토큰을 사용하지 않습니다. 대신 GPT-2는 바이트 페어 인코딩(BPE) 토크나이저를 사용해 단어를 부분단어로 분할합니다. 이에 대해서는 나중에 설명하겠습니다.

- 두 개의 독립된 텍스트 사이에 `<|endoftext|>` 토큰을 사용합니다:

<img src="https://sebastianraschka.com/images/LLMs-from-scratch-images/ch02_compressed/10.webp" width="800px">

- 다음과 같은 텍스트를 토큰화해 보죠:

````python
tokenizer = SimpleTokenizerV1(vocab)

text = "Hello, do you like tea. Is this-- a test?"

tokenizer.encode(text)
````

- 단어 "Hello"가 어휘사전에 없기 때문에 오류가 발생합니다.
- 이를 처리하기 위해 알지 못하는 단어를 표현하는 `"<|unk|>"` 같은 특수 토큰을 어휘 사전에 추가할 수 있습니다.
- 어휘사전을 준비했으므로 GPT-2 훈련에서 텍스트의 끝을 나타내기 위해 사용된 `"<|endoftext|>"` 토큰을 추가하겠습니다(또한 훈련 데이터셋이 여러 개의 글이나 책으로 구성된 경우 연결된 텍스트 사이를 나타내기 위해서도 사용됩니다).

````python
all_tokens = sorted(list(set(preprocessed)))
all_tokens.extend(["<|endoftext|>", "<|unk|>"])

vocab = {token:integer for integer,token in enumerate(all_tokens)}
````

````python
len(vocab.items())
````

````python
for i, item in enumerate(list(vocab.items())[-5:]):
    print(item)
````

- 새로운 `<unk>` 토큰을 사용할 때와 방법을 알려 주기 위해 토크나이저를 수정합니다.

````python
class SimpleTokenizerV2:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = { i:s for s,i in vocab.items()}

    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]
        preprocessed = [
            item if item in self.str_to_int
            else "<|unk|>" for item in preprocessed
        ]

        ids = [self.str_to_int[s] for s in preprocessed]
        return ids

    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        # 구둣점 문자 앞의 공백을 삭제합니다.
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
        return text
````

수정된 토크나이저로 텍스트를 토큰화해 보죠:

````python
tokenizer = SimpleTokenizerV2(vocab)

text1 = "Hello, do you like tea?"
text2 = "In the sunlit terraces of the palace."

text = " <|endoftext|> ".join((text1, text2))

print(text)
````

````python
tokenizer.encode(text)
````

````python
tokenizer.decode(tokenizer.encode(text))
````

## 2.5 바이트 페어 인코딩

- GPT-2는 바이트 페어 인코딩(BPE) 토크나이저를 사용합니다.
- 어휘사전에 없는 단어를 더 작은 부분단어나 개별 문자로 분할하여 처리할 수 있습니다.
- 예를 들어 GPT-2의 어휘사전에 단어 "unfamiliarword"가 없다면 이를 ["unfam", "iliar", "word"] 같이 토큰화할 수 있습니다. BPE의 훈련에 따라 결과가 달라질 수 있습니다.
- 원본 BPE 토크나이저는 다음 주소에 있습니다: [https://github.com/openai/gpt-2/blob/master/src/encoder.py](https://github.com/openai/gpt-2/blob/master/src/encoder.py)
- 이 장에서는 오픈AI의 오픈 소스 [tiktoken](https://github.com/openai/tiktoken) 라이브러리에서 제공하는 BPE 토크나이저를 사용합니다. 이 라이브러리는 계산 성능을 높이기 위해 핵심 알고리즘을 러스트(Rust)로 구현했습니다.
- [../02_bonus_bytepair-encoder](../02_bonus_bytepair-encoder) 폴더에 두 토크나이저 구현을 비교하는 노트북이 있습니다(tiktoken이 약 5배 빠릅니다).

````python
import tiktoken

tokenizer = tiktoken.get_encoding("gpt2")
````

# tiktoken의 역할과 최근 토크나이저 트렌드
`tiktoken`은 **텍스트를 LLM 입력용 토큰 ID로 빠르게 변환/복원**하는 라이브러리입니다. 이 예제에서는 GPT-2 방식(BPE) 토크나이징을 위해 사용됩니다.

- **고속 인코딩/디코딩**: 문자열 ↔ 토큰 ID 변환
- **모델별 인코딩 제공**: 예) `gpt2`, `cl100k_base`, `o200k_base`
- **특수 토큰 처리**: `<|endoftext|>` 같은 토큰 허용/제어
- **토큰 개수 계산**: 프롬프트 길이/비용 추정에 유용

이 노트북의 `tiktoken.get_encoding("gpt2")`는  
→ GPT-2 BPE 규칙을 불러와 텍스트를 토큰 ID로 바꾸는 설정입니다.

**최근(실무 기준)에는 무엇을 많이 쓰나?**

1. **OpenAI 계열 워크플로우**: `tiktoken`  
   - API 호출 전 토큰 수 계산, 입력 길이 관리에 많이 사용
2. **오픈소스 모델(Hugging Face 생태계)**: `transformers` + `tokenizers`  
   - `AutoTokenizer`가 사실상 표준
3. **Llama/T5/다국어 모델 계열**: `SentencePiece`(BPE/Unigram)  
   - 모델 배포본과 함께 제공되는 토크나이저를 그대로 사용하는 경우가 많음

정리하면,  
- **OpenAI 모델 중심**이면 `tiktoken`  
- **오픈소스 모델 학습/추론**이면 `AutoTokenizer`/`SentencePiece`  
가 현재 가장 일반적입니다.

- BPE 토크나이저는 알지 못하는 단어를 부분단어나 개별 문자로 분할합니다:

# BPE가 왜 “바이트 페어 인코딩”인지
BPE라는 이름은 원래 **데이터 압축 알고리즘**에서 왔고, 거기서 **자주 같이 나오는 바이트(byte) 쌍(pair)** 을 반복적으로 하나의 기호로 치환(encoding)했기 때문입니다.

## 예시 파일
[llm-from-scratch의 BPE 비교 노트북](https://github.com/rickiepark/llm-from-scratch/blob/main/ch02/02_bonus_bytepair-encoder/compare-bpe-tiktoken.ipynb)

1. **이름의 기원(압축 알고리즘 BPE)**  
   - 원래 BPE는 파일 압축에서 시작했습니다.  
   - 데이터에서 가장 자주 등장하는 **인접한 바이트 2개(바이트 페어)** 를 하나의 새 기호로 바꾸는 과정을 반복했습니다.  
   - 그래서 이름이 Byte-Pair-Encoding입니다.

2. **NLP에서의 BPE(서브워드 토크나이저)**  
   - 토큰화에서는 바이트/문자 단위에서 시작해, 자주 함께 나오는 **심볼 쌍**을 반복 병합합니다.  
   - 예: `l + o -> lo`, `lo + w -> low` 같은 식으로 점점 긴 서브워드를 만듭니다.  
   - 희귀 단어를 잘게 쪼개 처리할 수 있어 `<unk>`를 줄이는 데 유리합니다.

3. **왜 지금도 ‘바이트 페어’라고 부르나?**  
   - 역사적 이름이 그대로 남았기 때문입니다.  
   - 특히 GPT-2 계열처럼 **byte-level BPE**를 쓰는 경우는 이름이 기술적으로도 잘 맞습니다.  
   - 다만 일부 구현은 시작 단위를 문자/유니코드 심볼로 두기도 있어, 실무에서는 “쌍 병합 기반 서브워드 방식”으로 이해하면 가장 정확합니다.

- 데이터셋과 입력 텍스트 데이터셋에서 청크를 추출하는 데이터 로더를 만듭니다.

````python
import torch
# 파이토치 데이터 파이프라인의 핵심인 Dataset과 DataLoader 라이브러리를 가져옵니다.
# - Dataset: 커스텀 데이터셋을 만들 때 상속해야 하는 기본 클래스 (데이터의 길이와 값을 반환하는 기능 제공)
# - DataLoader: Dataset을 감싸서 미니배치 생성, 셔플링, 병렬 로딩 등을 지원하는 데이터 공급기
from torch.utils.data import Dataset, DataLoader


class GPTDatasetV1(Dataset):
    def __init__(self, txt, tokenizer, max_length, stride):
        """
        데이터셋을 초기화하고 입력 텍스트를 모델 학습에 맞는 텐서로 분할합니다.
        
        Args:
            txt (str): 전체 학습 텍스트 데이터
            tokenizer: 텍스트를 문자 단위에서 정수 ID로 변환해 줄 토크나이저 (예: 틱토큰)
            max_length (int): 모델이 한 번에 처리할 문맥의 최대 길이 (슬라이딩 윈도 크기)
            stride (int): 슬라이딩 윈도를 이동시킬 칸 수
        """
        self.input_ids = []
        self.target_ids = []

        # 1. 텍스트 전체를 한 번에 토큰화(정수 ID 리스트로 변환)합니다.
        # allowed_special: <|endoftext|> 등의 특수 토큰이 들어와도 에러 없이 처리하도록 허용합니다.
        token_ids = tokenizer.encode(txt, allowed_special={"<|endoftext|>"})
        
        # 입력 데이터가 다음 토큰을 예측하는 타깃 데이터를 만들 수 있을 만큼 충분히 긴지 검증합니다.
        assert len(token_ids) > max_length, "토큰화된 입력의 개수는 적어도 max_length+1과 같아야 합니다."

        # 2. 슬라이딩 윈도를 사용해 전체 텍스트를 max_length 길이의 예측 가능한 조각(청크)으로 나눕니다.
        # 한 번에 stride 만큼 점프하면서 시작 인덱스(i)를 잡습니다.
        for i in range(0, len(token_ids) - max_length, stride):
            # 모델의 입력 데이터 (예: [0, 1, 2])
            input_chunk = token_ids[i:i + max_length]
            # 모델이 맞춰야 할 정답 데이터 (입력보다 항상 1칸 뒤에 있는 토큰들, 예: [1, 2, 3])
            target_chunk = token_ids[i + 1: i + max_length + 1]
            
            # 3. 모델이 효율적으로 행렬 연산을 실행할 수 있도록 리스트를 텐서(torch.tensor)로 변환해 저장합니다.
            self.input_ids.append(torch.tensor(input_chunk))
            self.target_ids.append(torch.tensor(target_chunk))

    def __len__(self):
        """
        데이터셋의 전체 샘플 개수를 반환합니다.
        DataLoader가 이 데이터셋에서 만들어 낼 총 배치 개수를 계산할 때 내부적으로 호출합니다.
        """
        return len(self.input_ids)

    def __getitem__(self, idx):
        """
        특정 인덱스(idx) 번호에 해당하는 하나의 [입력 텐서, 타깃 텐서] 튜플을 반환합니다.
        DataLoader가 데이터를 반복(iter)하면서 하나씩 뽑아 배치를 구성할 때 지속적으로 호출되는 핵심 함수입니다.
        """
        return self.input_ids[idx], self.target_ids[idx]
````

````python
def create_dataloader_v1(txt, batch_size=4, max_length=256,
                         stride=128, shuffle=True, drop_last=True,
                         num_workers=0):

    # 토크나이저를 초기화합니다.
    tokenizer = tiktoken.get_encoding("gpt2")

    # 커스텀 데이터셋 인스턴스를 생성합니다.
    dataset = GPTDatasetV1(txt, tokenizer, max_length, stride)

    # 데이터 로더를 만듭니다. (각 파라미터는 모델 학습 시 데이터 공급 방식을 결정합니다.)
    dataloader = DataLoader(
        dataset,                 # 데이터를 가져올 Dataset 객체
        batch_size=batch_size,   # 한 번에 모델에 던져줄 샘플(미니배치)의 개수 (예: GPU 메모리에 맞게 조절)
        shuffle=shuffle,         # 에포크마다 데이터를 섞을지 여부 (학습 시에는 True 권장, 순서 암기 방지)
        drop_last=drop_last,     # 마지막 배치의 크기가 batch_size보다 작을 때, 그 배치를 버릴지(True) 사용할지 조절 (에러 방지용)
        num_workers=num_workers  # 데이터 로딩에 사용할 서브 프로세스 수 (0: 메인 프로세스만 사용, 멀티코어 환경에선 늘리면 로딩 속도 향상)
    )

    return dataloader
````

- 문맥 크기를 4와 배치 크기 1로 데이터 로더를 테스트해 보죠.:

# PyTorch DataLoader 파라미터 설명 및 사용법
`DataLoader`는 파이토치에서 데이터를 미니배치(Mini-batch)로 묶어주고 셔플링 같은 편의 기능을 제공하는 필수 클래스입니다. 코드에 주석을 달아 직관적으로 이해할 수 있도록 수정했습니다.

`DataLoader`를 생성할 때 사용하는 주요 파라미터는 모델의 학습 안정성과 속도에 큰 영향을 미칩니다. 각 역할은 다음과 같습니다.

1. **`dataset`**: (필수) 데이터를 가져올 출처입니다. 앞서 만든 `GPTDatasetV1` 객체처럼 인덱싱으로 데이터를 반환할 수 있어야 합니다.
2. **`batch_size`**: 모델이 한 번의 스텝(step)에서 처리할 데이터 개수입니다. 크면 GPU 메모리 에러(OOM)가 발생할 수 있고, 너무 작으면 학습 속도가 느려집니다.
3. **`shuffle`**: 매 에포크(Epoch) 시작 시 데이터를 섞을지 결정합니다. **학습(Train)** 중에는 `True`로 설정해 모델이 데이터 순서를 외우는 것을 방지하고, **검증(Validation)이나 테스트** 시에는 `False`로 진행합니다.
4. **`drop_last`**: 총 데이터 개수가 `batch_size`로 나누어떨어지지 않을 때, 남는 꼬투리 배치를 버릴지 설정합니다 (`True`시 버림). 일부 구조나 로스 계산 함수는 무조건 고정된 형태의 텐서를 원하기 때문에 이 값을 `True`로 주어 에러를 방지합니다. 
5. **`num_workers`**: 병렬 데이터 로딩에 사용할 코어(프로세스) 수입니다. 데이터 전처리(이미지 확대, 토큰화 등)가 복잡할 때 이 수치를 높이면 빠르지만 사용 환경 OS/메모리에 따라 한계가 있습니다. (`0`은 메인 스레드만 쓴다는 뜻입니다.)

## 2.8 단어 위치 인코딩하기

- 임베딩 층은 토큰 ID를 입력 시퀀스에서 어떤 위치에 있던지 상관없이 동일한 벡터 표현으로 바꿉니다:

<img src="https://sebastianraschka.com/images/LLMs-from-scratch-images/ch02_compressed/17.webp" width="600px">

- 위치 임베딩을 토큰 임베딩에 더해서 대규모 언어 모델을 위한 입력 임베딩을 만듭니다:

<img src="https://sebastianraschka.com/images/LLMs-from-scratch-images/ch02_compressed/18.webp" width="600px">

- 바이트 페어 인코더의 어휘 사전 크기는 50,257입니다.
- 입력 토큰을 256차원의 벡터 표현으로 인코딩한다고 가정합니다.

````python
vocab_size = 50257
output_dim = 256

token_embedding_layer = torch.nn.Embedding(vocab_size, output_dim)
````

- 데이터 로더에서 데이터를 샘플링한 다음 배치에 있는 각 샘플의 토큰을 256차원 벡터로 임베딩할 수 있습니다.
- 배치 크기가 8이고 샘플마다 4개의 토큰이 있다면 8 x 4 x 256 크기의 텐서가 만들어 집니다.

````python
max_length = 4
dataloader = create_dataloader_v1(
    raw_text, batch_size=8, max_length=max_length,
    stride=max_length, shuffle=False
)
data_iter = iter(dataloader)
inputs, targets = next(data_iter)
````

````python
print("토큰 ID:\n", inputs)
print("\n입력 크기:\n", inputs.shape)
````

````python
token_embeddings = token_embedding_layer(inputs)
print(token_embeddings.shape)

# 임베딩 벡터의 값을 확인합니다.
print(token_embeddings)
````

- GPT-2는 절대 위치 임베딩을 사용하므로 또 다른 임베딩 층을 만들면 됩니다:

````python
context_length = max_length
pos_embedding_layer = torch.nn.Embedding(context_length, output_dim)

# 임베딩 층의 가중치를 확인합니다.
print(pos_embedding_layer.weight)
````

````python
pos_embeddings = pos_embedding_layer(torch.arange(max_length))
print(pos_embeddings.shape)

# 위치 임베딩 값을 확인합니다.
print(pos_embeddings)
````

- LLM에 사용될 입력 임베딩을 만들기 위해 토큰 임베딩과 위치 임베딩을 더합니다:

````python
input_embeddings = token_embeddings + pos_embeddings
print(input_embeddings.shape)

# 입력 임베딩 값을 확인합니다.
print(input_embeddings)
````

- 입력 전처리 워크플로의 초기 단계에서 입력 텍스트를 별개의 토큰으로 분할합니다.
- 이 분할 단계 다음에 사전에 정의된 어휘사전을 기반으로 토큰을 토큰 ID로 변환합니다.
